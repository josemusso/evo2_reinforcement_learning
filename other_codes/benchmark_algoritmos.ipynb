# Data

```python
df_a2c = pd.read_csv(a2c+'progress.csv')
df_acer = pd.read_csv(acer+'progress.csv')
df_ppo2 = pd.read_csv(ppo2+'progress.csv')
df_trpo = pd.read_csv(trpo+'progress.csv')
df_ppo1 = pd.read_csv(ppo1+'progress.csv')
```

# Ep reward mean

## Data

```python
reward_mean_df = pd.DataFrame()
reward_mean_df['PPO1'] = df_ppo1['EpRewMean']
reward_mean_df['ACER'] = df_acer['mean_episode_reward']
reward_mean_df['TRPO'] = df_trpo['EpRewMean']

values = {"PPO1": df_ppo1['EpRewMean'].iloc[-1], "ACER": df_acer['mean_episode_reward'].iloc[-1], "TRPO": df_trpo['EpRewMean'].iloc[-1]}

reward_mean_df.fillna(value=values, inplace=True)
```

```python
reward_mean_df
```

## Graphics

```python
fig = px.line(reward_mean_df[['TRPO','ACER','PPO1']], title='Mean Episode Reward')
fig.update_xaxes(title_text='Iteration')
fig.update_yaxes(title_text='EpRewardMean')

fig.update_layout(
    autosize=False,
    width=1600,
    height=800,)

fig.show()
```

# Policy Entropy

in RL, entropy refers to the predictability of the actions of an agent.

## Data

```python
entropy = pd.DataFrame()
entropy['PPO1'] = df_ppo1['loss_ent']
entropy['PPO2'] = df_ppo2['policy_entropy']
entropy['ACER'] = df_acer['entropy']
entropy['A2C'] = df_a2c['policy_entropy']
entropy['TRPO'] = df_trpo['entropy']

values = {"PPO1": df_ppo1['loss_ent'].iloc[-1], "PPO2": df_ppo2['policy_entropy'].iloc[-1], "ACER": df_acer['entropy'].iloc[-1], "A2C": df_a2c['policy_entropy'].iloc[-1], "TRPO": df_trpo['entropy'].iloc[-1]}

entropy.fillna(value=values, inplace=True)
```

```python
entropy
```

## Graphics

```python
fig = px.line(entropy[['TRPO','ACER','PPO1','PPO2','A2C']], title='Entropy', log_y=True)
fig.update_xaxes(title_text='Iteration')
fig.update_yaxes(title_text='Entropy')

fig.update_layout(
    autosize=False,
    width=1600,
    height=800,)

fig.show()
```

# Explained Variance

The explained variance is used to measure the proportion of the variability of the predictions of a machine learning model. Simply put, it is the difference between the expected value and the predicted value.

## Data

```python
explained_variance = pd.DataFrame()
explained_variance['A2C'] = df_a2c['explained_variance']
explained_variance['PPO2'] = df_ppo2['explained_variance']
explained_variance['ACER'] = df_acer['explained_variance']
explained_variance['TRPO'] = df_trpo['explained_variance_tdlam_before']

values = {"A2C": df_a2c['explained_variance'].iloc[-1], "PPO2": df_ppo2['explained_variance'].iloc[-1], "ACER": df_acer['explained_variance'].iloc[-1], "TRPO": df_trpo['explained_variance_tdlam_before'].iloc[-1]}

explained_variance.fillna(value=values, inplace=True)
```

```python
explained_variance
```

## Graphics

```python
fig = px.line(explained_variance[['TRPO','ACER','PPO2','A2C']], title='Explained Variance')
fig.update_xaxes(title_text='Iteration')
fig.update_yaxes(title_text='Variance')

fig.update_layout(
    autosize=False,
    width=1600,
    height=800,)

fig.show()
```

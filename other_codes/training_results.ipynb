# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

# Data

```python
carpeta = '../benchmarks/10_tableros/'
compare_to = '../benchmarks/4_tableros/'
```

```python
df = pd.read_csv(carpeta+'progress.csv')
df_compare = pd.read_csv(compare_to+'progress.csv')

with open(carpeta+'current_actions_log.txt', 'r') as f:
    actions = f.read().splitlines()
```

```python
df
```

```python
df['time_elapsed_comp'] = df_compare['TimeElapsed']
df['EpLenMean_comp'] = df_compare['EpLenMean']
df['EpRewMean_comp'] = df_compare['EpRewMean']
```

# Graphs

```python
fig = px.violin(df, y=["TimeElapsed",'time_elapsed_comp'])

fig.update_xaxes(title_text='10 tableros vs 4 tableros')
fig.update_yaxes(title_text='Time')
fig.update_layout(
    autosize=False,
    width=1000,
    height=800,)

fig.show()
```

```python
fig = px.line(df, y=["TimeElapsed",'time_elapsed_comp'])
fig.update_layout(
    autosize=False,
    width=1000,
    height=800,)

fig.show()
```

## Mean Ep Length

```python
fig = px.line(df[['EpLenMean','EpLenMean_comp']], title='Mean Episode Length')
fig.update_xaxes(title_text='Iteration')
fig.update_yaxes(title_text='EpLenMean')

fig.add_hline(y=7, line_color='red')
fig.update_layout(
    autosize=False,
    width=1600,
    height=800,)

fig.show()
```

# Most Used premes

```python
d = {}
for action in actions:
    splited_action = action.replace('[', '').replace(']', '').replace(' ','').split(',')
    for splited in splited_action:
        if splited != '':
            number = int(splited)
            if number not in d:
                d[number] = 1
            else:
                d[number] += 1
    
```

```python
fig = px.bar(x=range(len(dict(sorted(d.items())))), 
            y=list(dict(sorted(d.items())).values()),  
            text_auto='.2s', 
            title="Frecuency of premes",
            labels=dict(x="Premes id", y="Frecuency")
            )
fig.update_xaxes(tickmode='linear')
fig.update_layout(
    autosize=False,
    width=1600,
    height=800,)
fig.show()
```
